```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('palmerpenguins_extended.csv')
df = df.dropna()  # drop rows with missing values
print(df.shape)
df.head()
```

```python
print(df.columns.tolist())
df.info()
```

```python
df.isnull().sum()
```

```python
# Numerical Features Only

features = [
    "bill_length_mm",
    "bill_depth_mm",
    "flipper_length_mm",
    "body_mass_g"
]

X = df[features]

print("Selected Features")
display(X.head())

print()

print("Shape :", X.shape)  
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# Convert into DataFrame

X_scaled = pd.DataFrame(
    X_scaled,
    columns=features
)

print("Scaled Dataset")

display(X_scaled.head())

print()

print(np.round(X_scaled.describe(),1))
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    X_scaled.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame(
    X_pca,
    columns=["PCA1","PCA2"]
)

print("PCA Dataset")

display(pca_df.head())

print()

print("Explained Variance")

print(pca.explained_variance_ratio_)
```

```python
wcss = []

for i in range(1,11):

    model = KMeans(
        n_clusters=i,
        random_state=42
    )

    model.fit(X_scaled)

    wcss.append(model.inertia_)

plt.figure(figsize=(8,5))

plt.plot(
    range(1,11),
    wcss,
    marker="o",
    linewidth=3
)

plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.grid(True)

plt.show()
```

```python
print("Silhouette Scores")
print()

for k in range(2,11):

    model = KMeans(
        n_clusters=k,
        random_state=42
    )

    labels = model.fit_predict(X_scaled)

    score = silhouette_score(
        X_scaled,
        labels
    )

    print(f"K = {k} ----> Score = {score:.4f}")
```

```python
kmeans = KMeans(
    n_clusters=3,
    random_state=42
)

clusters = kmeans.fit_predict(X_scaled)

df["Cluster"] = clusters

print("Cluster Added Successfully")

display(df.head())
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    pca_df["PCA1"],
    pca_df["PCA2"],
    c=df["Cluster"],
    s=60
)

plt.xlabel("PCA 1")
plt.ylabel("PCA 2")

plt.title("KMeans Clustering on Penguin Dataset")

plt.grid(True)

plt.show()
```

```python
print("Number of Penguins in Each Cluster")

print()

print(df["Cluster"].value_counts())

print()

sns.countplot(
    x="Cluster",
    data=df
)

plt.title("Cluster Distribution")

plt.show()
```
